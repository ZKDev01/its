```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

class SistemaConocimientoEstudiante:
    def __init__(self):
        """
        Sistema de Análisis de Conocimiento del Estudiante usando Lógica Difusa
        
        Variables de entrada:
        - Puntaje en exámenes (0-100)
        - Tiempo de estudio (0-20 horas)
        - Participación en clase (0-10)
        - Tareas completadas (0-100%)
        
        Variable de salida:
        - Nivel de conocimiento: No-Aprendido, Conocido, Insatisfecho, Aprendido
        """
        self.setup_variables()
        self.setup_membership_functions()
        self.setup_rules()
        self.create_control_system()
    
    def setup_variables(self):
        """Define las variables de entrada y salida del sistema"""
        # Variables de entrada
        self.puntaje_examen = ctrl.Antecedent(np.arange(0, 101, 1), 'puntaje_examen')
        self.tiempo_estudio = ctrl.Antecedent(np.arange(0, 21, 1), 'tiempo_estudio')
        self.participacion = ctrl.Antecedent(np.arange(0, 11, 1), 'participacion')
        self.tareas_completadas = ctrl.Antecedent(np.arange(0, 101, 1), 'tareas_completadas')
        
        # Variable de salida
        self.conocimiento = ctrl.Consequent(np.arange(0, 101, 1), 'conocimiento')
    
    def setup_membership_functions(self):
        """Define las funciones de membresía para cada variable"""
        
        # Puntaje en exámenes
        self.puntaje_examen['bajo'] = fuzz.trimf(self.puntaje_examen.universe, [0, 0, 40])
        self.puntaje_examen['medio'] = fuzz.trimf(self.puntaje_examen.universe, [30, 50, 70])
        self.puntaje_examen['alto'] = fuzz.trimf(self.puntaje_examen.universe, [60, 80, 100])
        self.puntaje_examen['excelente'] = fuzz.trimf(self.puntaje_examen.universe, [80, 90, 100])
        
        # Tiempo de estudio (horas por semana)
        self.tiempo_estudio['poco'] = fuzz.trimf(self.tiempo_estudio.universe, [0, 0, 5])
        self.tiempo_estudio['moderado'] = fuzz.trimf(self.tiempo_estudio.universe, [3, 8, 12])
        self.tiempo_estudio['intenso'] = fuzz.trimf(self.tiempo_estudio.universe, [10, 15, 20])
        
        # Participación en clase
        self.participacion['baja'] = fuzz.trimf(self.participacion.universe, [0, 0, 3])
        self.participacion['media'] = fuzz.trimf(self.participacion.universe, [2, 5, 8])
        self.participacion['alta'] = fuzz.trimf(self.participacion.universe, [7, 10, 10])
        
        # Tareas completadas (%)
        self.tareas_completadas['pocas'] = fuzz.trimf(self.tareas_completadas.universe, [0, 0, 40])
        self.tareas_completadas['algunas'] = fuzz.trimf(self.tareas_completadas.universe, [30, 60, 80])
        self.tareas_completadas['muchas'] = fuzz.trimf(self.tareas_completadas.universe, [70, 85, 100])
        
        # Nivel de conocimiento (salida)
        self.conocimiento['no_aprendido'] = fuzz.trimf(self.conocimiento.universe, [0, 0, 25])
        self.conocimiento['conocido'] = fuzz.trimf(self.conocimiento.universe, [15, 35, 55])
        self.conocimiento['insatisfecho'] = fuzz.trimf(self.conocimiento.universe, [45, 65, 85])
        self.conocimiento['aprendido'] = fuzz.trimf(self.conocimiento.universe, [75, 90, 100])
    
    def setup_rules(self):
        """Define las reglas del sistema de inferencia difusa"""
        self.rules = [
            # Reglas para "No Aprendido"
            ctrl.Rule(self.puntaje_examen['bajo'] & self.tiempo_estudio['poco'] & 
                     self.participacion['baja'] & self.tareas_completadas['pocas'], 
                     self.conocimiento['no_aprendido']),
            
            ctrl.Rule(self.puntaje_examen['bajo'] & self.tiempo_estudio['poco'], 
                     self.conocimiento['no_aprendido']),
            
            ctrl.Rule(self.participacion['baja'] & self.tareas_completadas['pocas'] & 
                     self.puntaje_examen['bajo'], 
                     self.conocimiento['no_aprendido']),
            
            # Reglas para "Conocido"
            ctrl.Rule(self.puntaje_examen['medio'] & self.tiempo_estudio['moderado'], 
                     self.conocimiento['conocido']),
            
            ctrl.Rule(self.participacion['media'] & self.tareas_completadas['algunas'] & 
                     self.puntaje_examen['medio'], 
                     self.conocimiento['conocido']),
            
            ctrl.Rule(self.tiempo_estudio['moderado'] & self.tareas_completadas['algunas'], 
                     self.conocimiento['conocido']),
            
            # Reglas para "Insatisfecho"
            ctrl.Rule(self.puntaje_examen['alto'] & self.tiempo_estudio['poco'], 
                     self.conocimiento['insatisfecho']),
            
            ctrl.Rule(self.participacion['alta'] & self.tareas_completadas['muchas'] & 
                     self.puntaje_examen['medio'], 
                     self.conocimiento['insatisfecho']),
            
            ctrl.Rule(self.tiempo_estudio['intenso'] & self.puntaje_examen['medio'], 
                     self.conocimiento['insatisfecho']),
            
            # Reglas para "Aprendido"
            ctrl.Rule(self.puntaje_examen['excelente'] & self.tiempo_estudio['intenso'] & 
                     self.participacion['alta'] & self.tareas_completadas['muchas'], 
                     self.conocimiento['aprendido']),
            
            ctrl.Rule(self.puntaje_examen['alto'] & self.tiempo_estudio['intenso'] & 
                     self.participacion['alta'], 
                     self.conocimiento['aprendido']),
            
            ctrl.Rule(self.puntaje_examen['excelente'] & self.tareas_completadas['muchas'], 
                     self.conocimiento['aprendido']),
            
            # Reglas adicionales para casos intermedios
            ctrl.Rule(self.puntaje_examen['bajo'] & self.tiempo_estudio['intenso'] & 
                     self.participacion['alta'], 
                     self.conocimiento['conocido']),
            
            ctrl.Rule(self.puntaje_examen['alto'] & self.participacion['baja'] & 
                     self.tareas_completadas['pocas'], 
                     self.conocimiento['insatisfecho']),
        ]
    
    def create_control_system(self):
        """Crea el sistema de control difuso"""
        self.sistema_ctrl = ctrl.ControlSystem(self.rules)
        self.simulacion = ctrl.ControlSystemSimulation(self.sistema_ctrl)
    
    def evaluar_estudiante(self, puntaje, tiempo, participacion, tareas):
        """
        Evalúa el nivel de conocimiento de un estudiante
        
        Args:
            puntaje: Puntaje en exámenes (0-100)
            tiempo: Tiempo de estudio en horas por semana (0-20)
            participacion: Nivel de participación (0-10)
            tareas: Porcentaje de tareas completadas (0-100)
        
        Returns:
            tuple: (valor_numerico, categoria_textual)
        """
        # Asignar valores de entrada
        self.simulacion.input['puntaje_examen'] = puntaje
        self.simulacion.input['tiempo_estudio'] = tiempo
        self.simulacion.input['participacion'] = participacion
        self.simulacion.input['tareas_completadas'] = tareas
        
        # Computar el resultado
        self.simulacion.compute()
        
        # Obtener el valor de salida
        valor_conocimiento = self.simulacion.output['conocimiento']
        
        # Clasificar en categorías
        categoria = self.clasificar_conocimiento(valor_conocimiento)
        
        return valor_conocimiento, categoria
    
    def clasificar_conocimiento(self, valor):
        """Clasifica el valor numérico en una categoría textual"""
        if valor <= 25:
            return "No-Aprendido"
        elif valor <= 50:
            return "Conocido"
        elif valor <= 80:
            return "Insatisfecho"
        else:
            return "Aprendido"
    
    def visualizar_variables(self):
        """Visualiza las funciones de membresía de las variables"""
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Funciones de Membresía - Sistema de Conocimiento del Estudiante', fontsize=16)
        
        # Puntaje examen
        self.puntaje_examen.view(ax=axes[0, 0])
        axes[0, 0].set_title('Puntaje en Exámenes')
        
        # Tiempo estudio
        self.tiempo_estudio.view(ax=axes[0, 1])
        axes[0, 1].set_title('Tiempo de Estudio (horas/semana)')
        
        # Participación
        self.participacion.view(ax=axes[0, 2])
        axes[0, 2].set_title('Participación en Clase')
        
        # Tareas completadas
        self.tareas_completadas.view(ax=axes[1, 0])
        axes[1, 0].set_title('Tareas Completadas (%)')
        
        # Conocimiento (salida)
        self.conocimiento.view(ax=axes[1, 1])
        axes[1, 1].set_title('Nivel de Conocimiento (Salida)')
        
        # Eliminar subplot vacío
        fig.delaxes(axes[1, 2])
        
        plt.tight_layout()
        plt.show()
    
    def generar_reporte_estudiante(self, nombre, puntaje, tiempo, participacion, tareas):
        """Genera un reporte completo del estudiante"""
        valor, categoria = self.evaluar_estudiante(puntaje, tiempo, participacion, tareas)
        
        print(f"\n{'='*50}")
        print(f"REPORTE DE ANÁLISIS DE CONOCIMIENTO")
        print(f"{'='*50}")
        print(f"Estudiante: {nombre}")
        print(f"{'='*50}")
        print(f"Variables de Entrada:")
        print(f"  • Puntaje en exámenes: {puntaje}/100")
        print(f"  • Tiempo de estudio: {tiempo} horas/semana")
        print(f"  • Participación en clase: {participacion}/10")
        print(f"  • Tareas completadas: {tareas}%")
        print(f"\nResultado del Análisis:")
        print(f"  • Valor numérico: {valor:.2f}/100")
        print(f"  • Categoría: {categoria}")
        print(f"{'='*50}")
        
        # Recomendaciones basadas en la categoría
        self.mostrar_recomendaciones(categoria, puntaje, tiempo, participacion, tareas)
        
        return valor, categoria
    
    def mostrar_recomendaciones(self, categoria, puntaje, tiempo, participacion, tareas):
        """Muestra recomendaciones basadas en el nivel de conocimiento"""
        print(f"\nRecomendaciones:")
        
        if categoria == "No-Aprendido":
            print("  🔴 Nivel crítico - Requiere intervención inmediata")
            print("  • Aumentar significativamente el tiempo de estudio")
            print("  • Buscar tutoría o apoyo académico adicional")
            print("  • Revisar métodos de estudio y comprensión de conceptos básicos")
            print("  • Mejorar participación en clase y completar todas las tareas")
            
        elif categoria == "Conocido":
            print("  🟡 Nivel básico - Progreso moderado")
            print("  • Mantener el ritmo actual de estudio")
            print("  • Incrementar gradualmente la participación en clase")
            print("  • Buscar oportunidades para profundizar en los temas")
            print("  • Completar más tareas para reforzar el aprendizaje")
            
        elif categoria == "Insatisfecho":
            print("  🟠 Buen potencial - Resultados inconsistentes")
            print("  • Optimizar métodos de estudio para mejores resultados en exámenes")
            print("  • Balancear el esfuerzo con los resultados obtenidos")
            print("  • Enfocarse en áreas específicas que necesitan mejora")
            print("  • Mantener el buen nivel de participación y tareas")
            
        else:  # Aprendido
            print("  🟢 Excelente nivel - Continuar con el buen trabajo")
            print("  • Mantener el alto rendimiento actual")
            print("  • Considerar desafíos adicionales o proyectos avanzados")
            print("  • Ayudar a compañeros que necesiten apoyo")
            print("  • Explorar temas relacionados para ampliar conocimientos")

# Función principal para demostrar el sistema
def main():
    # Crear el sistema
    sistema = SistemaConocimientoEstudiante()
    
    # Visualizar las funciones de membresía
    print("Generando visualización de funciones de membresía...")
    sistema.visualizar_variables()
    
    # Casos de prueba
    estudiantes = [
        ("Ana García", 25, 2, 1, 30),      # No-Aprendido
        ("Carlos López", 55, 8, 5, 65),    # Conocido
        ("María Rodríguez", 75, 4, 9, 85), # Insatisfecho
        ("Juan Pérez", 90, 15, 9, 95),     # Aprendido
        ("Sofia Chen", 45, 12, 7, 70),     # Caso intermedio
    ]
    
    print("\nAnalizando diferentes perfiles de estudiantes...\n")
    
    resultados = []
    for nombre, puntaje, tiempo, participacion, tareas in estudiantes:
        valor, categoria = sistema.generar_reporte_estudiante(
            nombre, puntaje, tiempo, participacion, tareas
        )
        resultados.append((nombre, valor, categoria))
        print("\n" + "="*50 + "\n")
    
    # Resumen final
    print("RESUMEN FINAL")
    print("="*50)
    for nombre, valor, categoria in resultados:
        print(f"{nombre:20} | {valor:6.2f} | {categoria}")
    
    return sistema

if __name__ == "__main__":
    sistema = main()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import pandas as pd

class SistemaConocimientoEstudianteMultidimensional:
    def __init__(self):
        """
        Sistema de Análisis de Conocimiento del Estudiante usando Lógica Difusa
        con evaluación multidimensional por áreas de conocimiento
        
        Variables de entrada por área:
        - Puntaje en exámenes (0-100)
        - Tiempo de estudio (0-20 horas)
        - Participación en clase (0-10)
        - Tareas completadas (0-100%)
        
        Áreas de conocimiento:
        - Álgebra
        - Análisis
        - Lógica Proposicional
        - Teoría de Conjuntos
        
        Salida: Tabla multidimensional con niveles por área
        """
        self.areas = ['Algebra', 'Analisis', 'Logica_Proposicional', 'Teoria_Conjuntos']
        self.niveles = ['No-Aprendido', 'Conocido', 'Insatisfecho', 'Aprendido']
        
        self.sistemas = {}
        self.simulaciones = {}
        
        # Crear un sistema difuso para cada área
        for area in self.areas:
            self.setup_variables_area(area)
            self.setup_membership_functions_area(area)
            self.setup_rules_area(area)
            self.create_control_system_area(area)
    
    def setup_variables_area(self, area):
        """Define las variables de entrada y salida para un área específica"""
        # Variables de entrada específicas para el área
        puntaje = ctrl.Antecedent(np.arange(0, 101, 1), f'puntaje_{area}')
        tiempo = ctrl.Antecedent(np.arange(0, 21, 1), f'tiempo_{area}')
        participacion = ctrl.Antecedent(np.arange(0, 11, 1), f'participacion_{area}')
        tareas = ctrl.Antecedent(np.arange(0, 101, 1), f'tareas_{area}')
        
        # Variable de salida para el área
        conocimiento = ctrl.Consequent(np.arange(0, 101, 1), f'conocimiento_{area}')
        
        if area not in self.sistemas:
            self.sistemas[area] = {}
        
        self.sistemas[area]['puntaje'] = puntaje
        self.sistemas[area]['tiempo'] = tiempo
        self.sistemas[area]['participacion'] = participacion
        self.sistemas[area]['tareas'] = tareas
        self.sistemas[area]['conocimiento'] = conocimiento
    
    def setup_membership_functions_area(self, area):
        """Define las funciones de membresía específicas para cada área"""
        sistema = self.sistemas[area]
        
        # Funciones de membresía ajustadas por área de conocimiento
        if area == 'Algebra':
            # Álgebra: más énfasis en práctica y resolución de problemas
            sistema['puntaje']['bajo'] = fuzz.trimf(sistema['puntaje'].universe, [0, 0, 45])
            sistema['puntaje']['medio'] = fuzz.trimf(sistema['puntaje'].universe, [35, 55, 75])
            sistema['puntaje']['alto'] = fuzz.trimf(sistema['puntaje'].universe, [65, 80, 95])
            sistema['puntaje']['excelente'] = fuzz.trimf(sistema['puntaje'].universe, [85, 95, 100])
            
        elif area == 'Analisis':
            # Análisis: requiere más tiempo de estudio y comprensión profunda
            sistema['puntaje']['bajo'] = fuzz.trimf(sistema['puntaje'].universe, [0, 0, 50])
            sistema['puntaje']['medio'] = fuzz.trimf(sistema['puntaje'].universe, [40, 60, 75])
            sistema['puntaje']['alto'] = fuzz.trimf(sistema['puntaje'].universe, [70, 85, 95])
            sistema['puntaje']['excelente'] = fuzz.trimf(sistema['puntaje'].universe, [90, 95, 100])
            
        elif area == 'Logica_Proposicional':
            # Lógica: más énfasis en participación y razonamiento
            sistema['puntaje']['bajo'] = fuzz.trimf(sistema['puntaje'].universe, [0, 0, 40])
            sistema['puntaje']['medio'] = fuzz.trimf(sistema['puntaje'].universe, [30, 50, 70])
            sistema['puntaje']['alto'] = fuzz.trimf(sistema['puntaje'].universe, [60, 75, 90])
            sistema['puntaje']['excelente'] = fuzz.trimf(sistema['puntaje'].universe, [80, 90, 100])
            
        else:  # Teoria_Conjuntos
            # Teoría de Conjuntos: balance entre teoría y práctica
            sistema['puntaje']['bajo'] = fuzz.trimf(sistema['puntaje'].universe, [0, 0, 42])
            sistema['puntaje']['medio'] = fuzz.trimf(sistema['puntaje'].universe, [32, 52, 72])
            sistema['puntaje']['alto'] = fuzz.trimf(sistema['puntaje'].universe, [62, 78, 92])
            sistema['puntaje']['excelente'] = fuzz.trimf(sistema['puntaje'].universe, [82, 92, 100])
        
        # Tiempo de estudio (ajustado por dificultad del área)
        if area in ['Analisis', 'Logica_Proposicional']:
            # Áreas más complejas requieren más tiempo
            sistema['tiempo']['poco'] = fuzz.trimf(sistema['tiempo'].universe, [0, 0, 6])
            sistema['tiempo']['moderado'] = fuzz.trimf(sistema['tiempo'].universe, [4, 9, 14])
            sistema['tiempo']['intenso'] = fuzz.trimf(sistema['tiempo'].universe, [12, 17, 20])
        else:
            # Álgebra y Teoría de Conjuntos
            sistema['tiempo']['poco'] = fuzz.trimf(sistema['tiempo'].universe, [0, 0, 5])
            sistema['tiempo']['moderado'] = fuzz.trimf(sistema['tiempo'].universe, [3, 8, 12])
            sistema['tiempo']['intenso'] = fuzz.trimf(sistema['tiempo'].universe, [10, 15, 20])
        
        # Participación (estándar para todas las áreas)
        sistema['participacion']['baja'] = fuzz.trimf(sistema['participacion'].universe, [0, 0, 3])
        sistema['participacion']['media'] = fuzz.trimf(sistema['participacion'].universe, [2, 5, 8])
        sistema['participacion']['alta'] = fuzz.trimf(sistema['participacion'].universe, [7, 10, 10])
        
        # Tareas completadas (ajustado por área)
        if area == 'Algebra':
            # Álgebra requiere más práctica
            sistema['tareas']['pocas'] = fuzz.trimf(sistema['tareas'].universe, [0, 0, 35])
            sistema['tareas']['algunas'] = fuzz.trimf(sistema['tareas'].universe, [25, 55, 75])
            sistema['tareas']['muchas'] = fuzz.trimf(sistema['tareas'].universe, [65, 80, 100])
        else:
            sistema['tareas']['pocas'] = fuzz.trimf(sistema['tareas'].universe, [0, 0, 40])
            sistema['tareas']['algunas'] = fuzz.trimf(sistema['tareas'].universe, [30, 60, 80])
            sistema['tareas']['muchas'] = fuzz.trimf(sistema['tareas'].universe, [70, 85, 100])
        
        # Nivel de conocimiento (salida) - ajustado por área
        sistema['conocimiento']['no_aprendido'] = fuzz.trimf(sistema['conocimiento'].universe, [0, 0, 25])
        sistema['conocimiento']['conocido'] = fuzz.trimf(sistema['conocimiento'].universe, [15, 35, 55])
        sistema['conocimiento']['insatisfecho'] = fuzz.trimf(sistema['conocimiento'].universe, [45, 65, 85])
        sistema['conocimiento']['aprendido'] = fuzz.trimf(sistema['conocimiento'].universe, [75, 90, 100])
    
    def setup_rules_area(self, area):
        """Define reglas específicas para cada área de conocimiento"""
        sistema = self.sistemas[area]
        
        # Reglas base comunes
        reglas_base = [
            # No Aprendido
            ctrl.Rule(sistema['puntaje']['bajo'] & sistema['tiempo']['poco'] & 
                     sistema['participacion']['baja'] & sistema['tareas']['pocas'], 
                     sistema['conocimiento']['no_aprendido']),
            
            ctrl.Rule(sistema['puntaje']['bajo'] & sistema['tiempo']['poco'], 
                     sistema['conocimiento']['no_aprendido']),
            
            # Conocido
            ctrl.Rule(sistema['puntaje']['medio'] & sistema['tiempo']['moderado'], 
                     sistema['conocimiento']['conocido']),
            
            ctrl.Rule(sistema['participacion']['media'] & sistema['tareas']['algunas'] & 
                     sistema['puntaje']['medio'], 
                     sistema['conocimiento']['conocido']),
            
            # Aprendido
            ctrl.Rule(sistema['puntaje']['excelente'] & sistema['tiempo']['intenso'] & 
                     sistema['participacion']['alta'] & sistema['tareas']['muchas'], 
                     sistema['conocimiento']['aprendido']),
            
            ctrl.Rule(sistema['puntaje']['alto'] & sistema['tiempo']['intenso'] & 
                     sistema['participacion']['alta'], 
                     sistema['conocimiento']['aprendido']),
        ]
        
        # Reglas específicas por área
        reglas_especificas = []
        
        if area == 'Algebra':
            # Álgebra: énfasis en práctica (tareas)
            reglas_especificas = [
                ctrl.Rule(sistema['tareas']['muchas'] & sistema['puntaje']['alto'], 
                         sistema['conocimiento']['aprendido']),
                ctrl.Rule(sistema['tareas']['pocas'] & sistema['puntaje']['alto'], 
                         sistema['conocimiento']['insatisfecho']),
                ctrl.Rule(sistema['tiempo']['poco'] & sistema['tareas']['muchas'], 
                         sistema['conocimiento']['conocido']),
            ]
            
        elif area == 'Analisis':
            # Análisis: énfasis en tiempo de estudio y comprensión
            reglas_especificas = [
                ctrl.Rule(sistema['tiempo']['intenso'] & sistema['puntaje']['medio'], 
                         sistema['conocimiento']['conocido']),
                ctrl.Rule(sistema['tiempo']['poco'] & sistema['puntaje']['alto'], 
                         sistema['conocimiento']['insatisfecho']),
                ctrl.Rule(sistema['tiempo']['intenso'] & sistema['participacion']['alta'], 
                         sistema['conocimiento']['aprendido']),
            ]
            
        elif area == 'Logica_Proposicional':
            # Lógica: énfasis en participación y razonamiento
            reglas_especificas = [
                ctrl.Rule(sistema['participacion']['alta'] & sistema['puntaje']['medio'], 
                         sistema['conocimiento']['conocido']),
                ctrl.Rule(sistema['participacion']['baja'] & sistema['puntaje']['alto'], 
                         sistema['conocimiento']['insatisfecho']),
                ctrl.Rule(sistema['participacion']['alta'] & sistema['tiempo']['moderado'], 
                         sistema['conocimiento']['aprendido']),
            ]
            
        else:  # Teoria_Conjuntos
            # Teoría de Conjuntos: balance equilibrado
            reglas_especificas = [
                ctrl.Rule(sistema['puntaje']['alto'] & sistema['participacion']['media'], 
                         sistema['conocimiento']['conocido']),
                ctrl.Rule(sistema['tiempo']['moderado'] & sistema['tareas']['algunas'], 
                         sistema['conocimiento']['conocido']),
                ctrl.Rule(sistema['puntaje']['medio'] & sistema['tiempo']['intenso'], 
                         sistema['conocimiento']['insatisfecho']),
            ]
        
        # Reglas para "Insatisfecho" (esfuerzo alto, resultados bajos)
        reglas_insatisfecho = [
            ctrl.Rule(sistema['tiempo']['intenso'] & sistema['puntaje']['bajo'], 
                     sistema['conocimiento']['insatisfecho']),
            ctrl.Rule(sistema['participacion']['alta'] & sistema['tareas']['muchas'] & 
                     sistema['puntaje']['medio'], 
                     sistema['conocimiento']['insatisfecho']),
        ]
        
        self.sistemas[area]['reglas'] = reglas_base + reglas_especificas + reglas_insatisfecho
    
    def create_control_system_area(self, area):
        """Crea el sistema de control difuso para un área específica"""
        sistema_ctrl = ctrl.ControlSystem(self.sistemas[area]['reglas'])
        self.simulaciones[area] = ctrl.ControlSystemSimulation(sistema_ctrl)
    
    def evaluar_estudiante_multidimensional(self, datos_areas):
        """
        Evalúa el nivel de conocimiento de un estudiante en todas las áreas
        
        Args:
            datos_areas: dict con estructura:
            {
                'Algebra': {'puntaje': 85, 'tiempo': 10, 'participacion': 7, 'tareas': 90},
                'Analisis': {'puntaje': 75, 'tiempo': 12, 'participacion': 8, 'tareas': 85},
                ...
            }
        
        Returns:
            pandas.DataFrame: Tabla con resultados por área y criterio
        """
        resultados = {}
        
        for area in self.areas:
            if area in datos_areas:
                datos = datos_areas[area]
                
                # Asignar valores de entrada
                self.simulaciones[area].input[f'puntaje_{area}'] = datos['puntaje']
                self.simulaciones[area].input[f'tiempo_{area}'] = datos['tiempo']
                self.simulaciones[area].input[f'participacion_{area}'] = datos['participacion']
                self.simulaciones[area].input[f'tareas_{area}'] = datos['tareas']
                
                # Computar resultado
                self.simulaciones[area].compute()
                
                # Obtener y clasificar resultado
                valor = self.simulaciones[area].output[f'conocimiento_{area}']
                categoria = self.clasificar_conocimiento(valor)
                
                resultados[area] = {
                    'Valor_Numerico': round(valor, 2),
                    'Categoria': categoria,
                    'Puntaje': datos['puntaje'],
                    'Tiempo': datos['tiempo'],
                    'Participacion': datos['participacion'],
                    'Tareas': datos['tareas']
                }
        
        # Crear DataFrame con los resultados
        df_resultados = pd.DataFrame(resultados).T
        df_resultados.index.name = 'Area'
        
        return df_resultados
    
    def clasificar_conocimiento(self, valor):
        """Clasifica el valor numérico en una categoría textual"""
        if valor <= 25:
            return "No-Aprendido"
        elif valor <= 50:
            return "Conocido"
        elif valor <= 80:
            return "Insatisfecho"
        else:
            return "Aprendido"
    
    def generar_tabla_conocimiento(self, datos_areas, nombre_estudiante="Estudiante"):
        """Genera una tabla completa de análisis de conocimiento"""
        df_resultados = self.evaluar_estudiante_multidimensional(datos_areas)
        
        print(f"\n{'='*80}")
        print(f"TABLA DE ANÁLISIS DE CONOCIMIENTO MULTIDIMENSIONAL")
        print(f"Estudiante: {nombre_estudiante}")
        print(f"{'='*80}")
        
        # Mostrar tabla principal
        print("\nRESULTADOS POR ÁREA:")
        print("-" * 80)
        formato_tabla = "{:<20} {:<12} {:<15} {:<8} {:<8} {:<8} {:<8}"
        print(formato_tabla.format("ÁREA", "CATEGORÍA", "VALOR", "PUNTAJE", "TIEMPO", "PARTIC.", "TAREAS"))
        print("-" * 80)
        
        for area in df_resultados.index:
            row = df_resultados.loc[area]
            area_nombre = area.replace('_', ' ').title()
            print(formato_tabla.format(
                area_nombre,
                row['Categoria'],
                f"{row['Valor_Numerico']}/100",
                f"{row['Puntaje']}/100",
                f"{row['Tiempo']}h",
                f"{row['Participacion']}/10",
                f"{row['Tareas']}%"
            ))
        
        # Análisis estadístico
        self.mostrar_analisis_estadistico(df_resultados)
        
        # Recomendaciones por área
        self.mostrar_recomendaciones_multidimensionales(df_resultados)
        
        return df_resultados
    
    def mostrar_analisis_estadistico(self, df_resultados):
        """Muestra análisis estadístico de los resultados"""
        print(f"\n{'='*40}")
        print("ANÁLISIS ESTADÍSTICO")
        print(f"{'='*40}")
        
        # Contar categorías
        conteo_categorias = df_resultados['Categoria'].value_counts()
        print("Distribución por nivel:")
        for categoria, cantidad in conteo_categorias.items():
            print(f"  • {categoria}: {cantidad} área(s)")
        
        # Promedios
        promedio_valor = df_resultados['Valor_Numerico'].mean()
        promedio_puntaje = df_resultados['Puntaje'].mean()
        promedio_tiempo = df_resultados['Tiempo'].mean()
        
        print(f"\nPromedios generales:")
        print(f"  • Valor de conocimiento: {promedio_valor:.2f}/100")
        print(f"  • Puntaje en exámenes: {promedio_puntaje:.2f}/100")
        print(f"  • Tiempo de estudio: {promedio_tiempo:.2f} horas/semana")
        
        # Área con mejor/peor rendimiento
        mejor_area = df_resultados.loc[df_resultados['Valor_Numerico'].idxmax()]
        peor_area = df_resultados.loc[df_resultados['Valor_Numerico'].idxmin()]
        
        print(f"\nRendimiento por área:")
        print(f"  • Mejor área: {mejor_area.name.replace('_', ' ').title()} ({mejor_area['Valor_Numerico']}/100)")
        print(f"  • Área a mejorar: {peor_area.name.replace('_', ' ').title()} ({peor_area['Valor_Numerico']}/100)")
    
    def mostrar_recomendaciones_multidimensionales(self, df_resultados):
        """Muestra recomendaciones específicas por área"""
        print(f"\n{'='*40}")
        print("RECOMENDACIONES POR ÁREA")
        print(f"{'='*40}")
        
        for area in df_resultados.index:
            row = df_resultados.loc[area]
            area_nombre = area.replace('_', ' ').title()
            
            print(f"\n📚 {area_nombre}:")
            print(f"   Estado: {row['Categoria']} ({row['Valor_Numerico']}/100)")
            
            if row['Categoria'] == "No-Aprendido":
                print("   🔴 Requiere atención urgente")
                if area == 'Algebra':
                    print("   • Aumentar práctica con ejercicios básicos")
                    print("   • Reforzar conceptos fundamentales")
                elif area == 'Analisis':
                    print("   • Incrementar tiempo de estudio significativamente")
                    print("   • Buscar tutoría especializada")
                elif area == 'Logica_Proposicional':
                    print("   • Participar más activamente en discusiones")
                    print("   • Practicar razonamiento lógico diariamente")
                else:  # Teoria_Conjuntos
                    print("   • Revisar fundamentos teóricos")
                    print("   • Practicar con ejemplos concretos")
                    
            elif row['Categoria'] == "Conocido":
                print("   🟡 Progreso estable, continuar mejorando")
                if area == 'Algebra':
                    print("   • Aumentar complejidad de ejercicios")
                elif area == 'Analisis':
                    print("   • Profundizar en demostraciones")
                elif area == 'Logica_Proposicional':
                    print("   • Explorar aplicaciones avanzadas")
                else:
                    print("   • Conectar con otras áreas matemáticas")
                    
            elif row['Categoria'] == "Insatisfecho":
                print("   🟠 Optimizar métodos de estudio")
                print("   • Revisar eficiencia del tiempo invertido")
                print("   • Cambiar estrategias de aprendizaje")
                
            else:  # Aprendido
                print("   🟢 Excelente dominio")
                print("   • Mantener el nivel actual")
                print("   • Considerar temas avanzados")

    def visualizar_tabla_heatmap(self, df_resultados, nombre_estudiante="Estudiante"):
        """Crea un heatmap visual de los resultados"""
        # Preparar datos para el heatmap
        areas_nombres = [area.replace('_', ' ').title() for area in df_resultados.index]
        valores = df_resultados['Valor_Numerico'].values.reshape(1, -1)
        
        # Crear el heatmap
        fig, ax = plt.subplots(figsize=(12, 4))
        im = ax.imshow(valores, cmap='RdYlGn', aspect='auto', vmin=0, vmax=100)
        
        # Configurar etiquetas
        ax.set_xticks(range(len(areas_nombres)))
        ax.set_xticklabels(areas_nombres, rotation=45, ha='right')
        ax.set_yticks([0])
        ax.set_yticklabels([nombre_estudiante])
        
        # Añadir valores en las celdas
        for i in range(len(areas_nombres)):
            categoria = df_resultados.iloc[i]['Categoria']
            valor = df_resultados.iloc[i]['Valor_Numerico']
            ax.text(i, 0, f'{categoria}\n{valor:.1f}', 
                   ha='center', va='center', fontweight='bold')
        
        # Colorbar
        cbar = plt.colorbar(im, ax=ax)
        cbar.set_label('Nivel de Conocimiento (0-100)', rotation=270, labelpad=20)
        
        plt.title(f'Mapa de Conocimiento - {nombre_estudiante}', fontsize=14, pad=20)
        plt.tight_layout()
        plt.show()

# Función principal para demostrar el sistema
def main():
    # Crear el sistema multidimensional
    sistema = SistemaConocimientoEstudianteMultidimensional()
    
    # Casos de prueba
    estudiantes_datos = {
        "Ana García": {
            'Algebra': {'puntaje': 30, 'tiempo': 3, 'participacion': 2, 'tareas': 40},
            'Analisis': {'puntaje': 25, 'tiempo': 2, 'participacion': 1, 'tareas': 30},
            'Logica_Proposicional': {'puntaje': 35, 'tiempo': 4, 'participacion': 3, 'tareas': 45},
            'Teoria_Conjuntos': {'puntaje': 28, 'tiempo': 3, 'participacion': 2, 'tareas': 35}
        },
        "Carlos López": {
            'Algebra': {'puntaje': 65, 'tiempo': 8, 'participacion': 6, 'tareas': 75},
            'Analisis': {'puntaje': 55, 'tiempo': 10, 'participacion': 5, 'tareas': 70},
            'Logica_Proposicional': {'puntaje': 70, 'tiempo': 7, 'participacion': 7, 'tareas': 80},
            'Teoria_Conjuntos': {'puntaje': 60, 'tiempo': 9, 'participacion': 6, 'tareas': 72}
        },
        "María Rodríguez": {
            'Algebra': {'puntaje': 70, 'tiempo': 5, 'participacion': 9, 'tareas': 85},
            'Analisis': {'puntaje': 65, 'tiempo': 6, 'participacion': 8, 'tareas': 90},
            'Logica_Proposicional': {'puntaje': 75, 'tiempo': 4, 'participacion': 10, 'tareas': 95},
            'Teoria_Conjuntos': {'puntaje': 68, 'tiempo': 5, 'participacion': 9, 'tareas': 88}
        },
        "Juan Pérez": {
            'Algebra': {'puntaje': 92, 'tiempo': 15, 'participacion': 9, 'tareas': 98},
            'Analisis': {'puntaje': 88, 'tiempo': 18, 'participacion': 10, 'tareas': 95},
            'Logica_Proposicional': {'puntaje': 95, 'tiempo': 12, 'participacion': 10, 'tareas': 100},
            'Teoria_Conjuntos': {'puntaje': 90, 'tiempo': 14, 'participacion': 9, 'tareas': 96}
        }
    }
    
    print("Analizando perfiles multidimensionales de estudiantes...\n")
    
    resultados_todos = {}
    
    for nombre, datos in estudiantes_datos.items():
        df_resultado = sistema.generar_tabla_conocimiento(datos, nombre)
        resultados_todos[nombre] = df_resultado
        
        # Generar heatmap visual
        sistema.visualizar_tabla_heatmap(df_resultado, nombre)
        
        print("\n" + "="*80 + "\n")
    
    return sistema, resultados_todos

if __name__ == "__main__":
    sistema, resultados = main()
```
